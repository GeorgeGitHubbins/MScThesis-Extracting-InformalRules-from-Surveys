```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_IN = pd.read_csv("SCALAR_Coastal_Longitudinal_Study_Wave_One_IN.csv")
df_USA = pd.read_csv("SCALAR_Coastal_Longitudinal_Study_Wave_One_USA.csv")
```

```python
df_Jakarta = df_IN[(df_IN["Postcode"] >= 10000) & (df_IN["Postcode"] < 20000)]
df_Jakarta = df_Jakarta.set_index("RecordNo")
df_Jakarta
```

```python
df_Houston = df_USA[(df_USA["Zipcode"] > 77000) & (df_USA["Zipcode"] < 78000)]
df_Houston = df_Houston.set_index("RecordNo")
df_Houston
```

```python
Qs_Jakarta = ["Gender", "Age", "Ethnicity", "Education",
              "Employment", "Postcode"]
Qs_Houston = ["Gender", "Age", "Race", "Education",
              "Employment", "State", "Zipcode"]
Qs_both = ["Q1", "Q5", "Q10",
           "Q13a", "Q13b", "Q13c", "Q13d", "Q13e", "Q13f",
           "Q47a", "Q47b", "Q52", "Q52a", "Q53", "Q54",
           "Q55", "Q55c", "Q56", "Q56b", "Q58", "Q59",
           "Q60a_1", "Q60a_2", "Q60a_3", "Q60a_99", "Q61"]

df_Houston = df_Houston[Qs_Houston + Qs_both]
df_Jakarta = df_Jakarta[Qs_Jakarta + Qs_both]
```

```python
df_Jakarta.to_csv("Survey_Wave_One_Jakarta_subset.csv")
df_Houston.to_csv("Survey_Wave_One_Houston_subset.csv")
```
